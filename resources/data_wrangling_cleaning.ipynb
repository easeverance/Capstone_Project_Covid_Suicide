```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
socio_data = pd.read_csv("us_county_sociohealth_data.csv")
```

```python
ed_data = pd.read_csv("education.csv")
```

```python
socio_df = pd.DataFrame(socio_data)
```

```python
ed_df = pd.DataFrame(ed_data)
```

```python
socio_df
```

```python
ed_df
```

```python
ed_df = ed_df[["FIPS Code", "State", "Area name", "Percent of adults completing some college or associate's degree, 2000", "Percent of adults with a bachelor's degree or higher, 2000", "Less than a high school diploma, 2015-19", "High school diploma only, 2015-19", "Some college or associate's degree, 2015-19", "Bachelor's degree or higher, 2015-19", "Percent of adults with less than a high school diploma, 2015-19", "Percent of adults with a high school diploma only, 2015-19", "Percent of adults completing some college or associate's degree, 2015-19", "Percent of adults with a bachelor's degree or higher, 2015-19"]]
```

```python
ed_df
```

```python
ed_df = ed_df.rename(columns={"FIPS Code": "fips"})
ed_df = ed_df.rename(columns={"State": "state"})
ed_df = ed_df.rename(columns={"Area name": "county"})
ed_df = ed_df.rename(columns={"Percent of adults completing some college or associate's degree, 2000": "some_college_or_associates_2000"})
ed_df = ed_df.rename(columns={"Percent of adults with a bachelor's degree or higher, 2000": "bachelors_or_higher_2000"})
ed_df = ed_df.rename(columns={"Less than a high school diploma, 2015-19": "less_than_highschool_2015_19"})
ed_df = ed_df.rename(columns={"High school diploma only, 2015-19": "highschool_only_2015_19"})
ed_df = ed_df.rename(columns={"Some college or associate's degree, 2015-19": "some_college_or_associates_2015_19"})
ed_df = ed_df.rename(columns={"Bachelor's degree or higher, 2015-19": "bachelors_or_higher_2015_19"})
ed_df = ed_df.rename(columns={"Percent of adults with less than a high school diploma, 2015-19": "percent_less_than_highschool_2015_19"})
ed_df = ed_df.rename(columns={"Percent of adults with a high school diploma only, 2015-19": "percent_highschool_only_2015_19"})
ed_df = ed_df.rename(columns={"Percent of adults completing some college or associate's degree, 2015-19": "percent_some_college_or_associates_2015_19"})
ed_df = ed_df.rename(columns={"Percent of adults with a bachelor's degree or higher, 2015-19": "percent_bachelors_or_higher_2015_19"})
```
